```python
from glob2 import glob
json_files = glob(sources+'/**/*/*.json') #Can be used absolute or relative paths
```

```python
source_links = []
for source in noUS_json_files:
    df2 = pd.read_json(source, orient='index')
    y = [df2.loc['newspaper'][0][i]['website'] for i in range(len(df2.loc['newspaper'][0]))]
    source_links += y
```

```python
def remove_http(row):
    row[0] = re.sub('http\:\/\/|https\:\/\/|\/.*|\?.*|\#.*','', row[0])
```

```python
# load chrome data
browser_history_path = '/Users/snorrealvsvag/MediaFutures/codeprojects/showing-news-wp1/backend/data/Takeout/Chrome/BrowserHistory.json'

browser_history = pd.read_json(browser_history_path)
```

```python
browser_history = pd.json_normalize(browser_history['Browser History'])
browser_history['link'] = None
```

```python
browser_history
```

```python
def add_source_url(row):
    source_url = re.sub('http\:\/\/|https\:\/\/|\/.*|\?.*|\#.*','', row.url)
    row.link = source_url
    return row
    
browser_history = browser_history.apply(add_source_url, axis=1)
```
